```python
import matplotlib.pyplot as plt
import numpy as np


Fs = 1000
f = 150
x = np.linspace(0, 100, 1000)
hight = np.cos(x / 10) * np.exp(-x / 20) + 1
plt.figure(figsize=(12, 3))
plt.plot(x, hight, "r")
y = hight * np.sin(2 * np.pi * f * x / Fs)
plt.plot(x, abs(y), "black")
plt.axhline(y=1, linestyle="--")
plt.xlabel("Time(s)")
plt.ylabel("Ventialtion")
plt.show()
```
